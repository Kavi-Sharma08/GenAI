```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([10,20,30,40,50,1000 ,1001,1008,1100,1004,1209,1210])
```

```python
def min_max_scaling(data):
    min_val = np.min(data)
    max_val = np.max(data)
    return (data - min_val)/(max_val - min_val)
```

```python
def z_score_standardization(data):
    mean = np.mean(data)
    std_deviation = np.std(data)
    return (data - mean) / std_deviation
```

```python
min_max_scaled = min_max_scaling(data)
z_score = z_score_standardization(data)
plt.scatter(range(len(min_max_scaled)), min_max_scaled)
plt.title('Min max Scaling Plot')
```

```python
print("Original Data " , data)
print("Min Max Scaling " , min_max_scaling(data))
print("z_score_standarization " , z_score_standardization(data))
```

```python
plt.scatter(range(len(z_score)) , z_score)
plt.title('Z Score Standardization')
```
